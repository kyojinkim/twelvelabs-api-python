## Variables

```python
api_key = ""
```

## Search

```python
index_id = ""

search_query = ""
search_options = ["visual", "conversation"]
```

```python
from twelvelabs import TwelveLabs
from twelvelabs.models.search import SearchData

import time

# Initialize Client
client = TwelveLabs(api_key=api_key)

# Keyword Search
response = client.search.query(
    index_id=index_id,
    query_text=search_query,
    options=search_options,
    group_by="clip",
    threshold="low",
    page_limit=12
)

# Utility function to print a specific page
def show_response(data: SearchData):
    start_time = time.time()
    for item in data:
        print(f"    video_id={item.video_id} score={item.score} start={item.start} end={item.end} confidence={item.confidence}")

    end_time = time.time()
    taken_time = end_time - start_time
    print(f"took {taken_time} seconds")

show_response(response.data)

while True:
    try:
        show_response(next(response))
    except StopIteration:
        break
```
