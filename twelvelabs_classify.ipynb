## Indexes

classes for classify should be defined directly in the code.

unable to use the classes with group which is existed in Playground.

```python
index_id = ""

classify_options = ["visual", "conversation"]
classify_conversation_option = "semantic" # ["semantic", "exact_match"]
classify_threshold = None
classify_classes = [
  {
    "name": "Coffee",
    "prompts": [
      "brewing",
      "pour coffee",
      "choose beans",
      "enjoy coffee",
      "small talk"
    ]
  },
  {
    "name": "Tea",
    "prompts": [
      "order tea",
      "smell tea",
      "choose tea",
      "pour tea"
    ]
  }
]
```

```python
from twelvelabs import TwelveLabs
from twelvelabs.models.classify import ClassifyVideoData

import time

# Initialize Client
client = TwelveLabs(api_key=api_key)

# Classify
response = client.classify.index(
    index_id=index_id,
    options=classify_options,
    classes=classify_classes,
    conversation_option=classify_conversation_option,
    page_limit=12,
    include_clips=True,
    threshold=classify_threshold,
    show_detailed_score=True
)

# Utility function to print a specific page
def show_response(data: ClassifyVideoData):
    start_time = time.time()
    for item in data:
        for _class in item.classes:
            for clip in _class.clips:
                print(f"    video_id={item.video_id} class={_class.name} score={_class.score} duration_ratio={_class.duration_ratio} start={clip.start} end={clip.end} thumbnail={clip.thumbnail_url}")

    end_time = time.time()
    taken_time = end_time - start_time
    print(f"took {taken_time} seconds")

show_response(response.data)

while True:
    try:
        show_response(next(response))
    except StopIteration:
        break
```
